```python
from sklearn.feature_extraction.text import *
f = open('dim_items.txt')
```

```python
line = f.readlines()
list = []
```

```python
for each in line:
    fc = each.split()
    fc = fc[2]    
    fc = fc.replace(',', ' ')
    list.append(fc)
```

```python
list
```

```python
vector = CountVectorizer()
res = vector.fit_transform(list)
```

```python
res.toarray()
```

```python
transformer = TfidfTransformer()
tfidf = transformer.fit_transform(res)
```

```python
tfidf.toarray()
```
